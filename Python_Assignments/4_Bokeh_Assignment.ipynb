Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Answer =
Glyphs are simple shapes used to represent data points visually in a plot. They can be different shapes  custom shape defined by the user. Each glyph is associated with a data source, which provides the coordinates of the points to be plotted.

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show,output_file
from bokeh.models import ColumnDataSource
output_file('test.html')
x = [1, 2, 3, 4, 5]
y = [2, 5, 4, 6, 3]
p=figure(title='samples')
p.xaxis.axis_label="x"
p.yaxis.axis_label="y"
p.circle(x,y)
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?


```python
x = [3,4,5,6,7]
y = [4,5,6,7,4]
output_file("line.html")
p = figure(title = "line graph")
p.scatter(x,y , size = 20,fill_color = 'red',legend_label = 'this is my data point')
show(p)
```
