```python
#Q1:
class Vehicle:
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle):
        self.name_of_vehicle = name_of_vehicle
        self.max_speed = max_speed
        self.average_of_vehicle = average_of_vehicle

#Q2:
class Car(Vehicle):
    def __init__(self, name_of_vehicle, max_speed, average_of_vehicle, seating_capacity):
        super().__init__(name_of_vehicle, max_speed, average_of_vehicle)
        self.seating_capacity = seating_capacity

    def seating_capacity(self, capacity):
        return f"{self.name_of_vehicle} has a seating capacity of {capacity}"

#Q3:#Multiple inheritance is a feature of object-oriented programming that allows a class to inherit from multiple parent classes. This can be useful for creating classes that share a variety of characteristics. However, it can also be complex and lead to ambiguity, so it is important to use multiple inheritance carefully.


class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def make_sound(self):
        pass

class Dog(Animal):
    def __init__(self, name, species, breed):
        super().__init__(name, species)
        self.breed = breed

    def make_sound(self):
        print("Woof!")

class Cat(Animal):
    def __init__(self, name, species, color):
        super().__init__(name, species)
        self.color = color

    def make_sound(self):
        print("Meow!")

class SiameseCat(Cat):
    def __init__(self, name, color):
        super().__init__(name, "Siamese", color)

siamese_cat = SiameseCat("Luna", "white")
siamese_cat.make_sound()

#Q4: Getters and setters are methods that allow you to get and set the values of private attributes of a class. Getters are methods that return the value of an attribute, while setters are methods that set the value of an attribute.
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age

    def get_name(self):
        return self._name

    def set_name(self, new_name):
        self._name = new_name

person = Person("Alice", 25)
name = person.get_name()
person.set_name("Bob")
print(person.get_name())

#Q5:Method overriding is a feature of object-oriented programming that allows a child class to redefine a method that is inherited from a parent class. This can be useful for customizing the behavior of a method for a specific class.


class Animal:
    def make_sound(self):
        print("Animal sound")

class Dog(Animal):
    def make_sound(self):
        print("Woof!")

dog = Dog()
dog.make_sound()




In this example, the `Dog` class overrides the `make_sound()` method that is inherited from the `Animal` class. This means that when the `make_sound()` method is called on a `Dog` object, it will print "Woof!", instead of "Animal sound".
```
