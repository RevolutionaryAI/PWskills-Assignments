Answers:-

Q1. MongoDB is a NoSQL, document-oriented database that stores data in flexible, semi-structured BSON format. Non-relational databases like MongoDB are preferred for handling unstructured or rapidly changing data, high scalability, and horizontal scaling, making them suitable for scenarios like real-time analytics, content management, and social media applications.

Q2. MongoDB features:
a. Schema-less: No fixed schema, flexibility in data structure.
b. High Scalability: Supports horizontal scaling across multiple servers.
c. Rich Query Language: Provides powerful querying capabilities.
d. Replication and Fault Tolerance: Data redundancy for reliability.
e. Sharding: Distributes data across clusters for better performance.
f. Indexing: Allows efficient data retrieval.
g. Geospatial Capabilities: Supports geospatial data and queries.

Q7:-

delete_one(): Deletes the first document that matches a query.
delete_many(): Deletes all documents that match a query.
drop(): Drops the entire collection, removing all documents and the collection itself.
