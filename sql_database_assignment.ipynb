Q1. What is a database? Differentiate between SQL and NoSQL databases.

A database is a system for storing and organizing data. It is a collection of data that is organized in a way that makes it easy to access, manage, and update.

There are two main types of databases: SQL and NoSQL.

SQL databases are relational databases. This means that they store data in tables that are related to each other. SQL databases are typically used for structured data, such as customer information, product information, and financial data.

NoSQL databases are non-relational databases. This means that they do not store data in tables. NoSQL databases are typically used for unstructured data, such as images, videos, and social media posts.

Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

DML stands for Data Manipulation Language. DML is used to manipulate data in a database. DML statements are used to insert, update, and delete data in tables.

UPDATE: Updates existing data in a table. For example, the following SQL statement updates the email address of an existing customer in the customers table:

UPDATE customers SET email = 'jane.doe@example.com' WHERE id = 1;

DELETE: Deletes rows from a table. For example, the following SQL statement deletes the customer with the ID of 1 from the customers table:

DELETE FROM customers WHERE id = 1;

Q4. What is DQL? Explain SELECT with an example.

DQL stands for Data Query Language. DQL is used to query data from a database. DQL statements are used to select data from tables and to perform calculations on the data.

The following is an example of a DQL statement:

SELECT * FROM customers;

Q5. Explain Primary Key and Foreign Key.

A primary key is a column or set of columns in a table that uniquely identifies each row in the table. A foreign key is a column or set of columns in a table that references the primary key of another table.

Primary keys are used to enforce data integrity and to ensure that each row in a table is unique. Foreign keys are used to create relationships between tables.

For example, the customers table might have a primary key column called id. The orders table might have a foreign key column called customer_id that references the id column in the customers table. This relationship indicates that each order in the orders table belongs to a customer in the customers table.
