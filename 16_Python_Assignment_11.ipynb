This means that we can use the try, except, and finally statements to handle our custom exception in the same way that we would handle any other exception.

```python
try:
    # Code that may raise a MyCustomException
except MyCustomException as e:
    # Handle the exception
```

Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

The following errors are defined in the ArithmeticError class:

FloatingPointError: Raised when an operation results in a floating-point overflow or underflow.

OverflowError: Raised when an operation results in a numeric overflow.

ZeroDivisionError: Raised when an attempt is made to divide by zero.

```python
#Example of FloatingPointError:
try:
    10.0 / 0.0
except FloatingPointError:
    print("Floating-point error")
    
#Example of OverflowError:
try:
    1000000000000000000000000000000000000000
except OverflowError:
    print("Overflow error")

#Example of ZeroDivisionError:
try:
    10 / 0
except ZeroDivisionError:
    print("Zero division error")
```
