```python
"""Q4. Why self is used in OOPs?
--> self is used for the instance of the class.It is reference to the object and used
to access the object's attributes and methods from within the class."""

class Car:
  def __init__(self,make,model):
    self.make=make
    self.model=model

  def print_details(self):
        print("Make:", self.make)
        print("Model:", self.model)

car1=Car("Honda","Civic")
car1.print_details()
```

```python
'''Q5. What is inheritance? Give an example for each type of inheritance.
->It is used in OOPs to allow a class to inherit its property and methods of another class.
The class that inherits properties and methos is called class or subclass,and the
class in which it inherits is called parent class or superclass.

1.Single Inheritance-In single inheritance, a subclass inherits from a single parent class.
'''
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def print_details(self):
        print("Make:", self.make)
        print("Model:", self.model)

class Car(Vehicle):
    pass

car1 = Car("Honda", "Civic")
print(car1.print_details())

'''Multiple Inheritance: In multiple inheritance, a subclass inherits from multiple parent classes.'''
class A:
  def methodA(self):
    print("Method A")

class B:
  def methodB(self):
    print("Method B")

class C(A,B):
  pass
obj=C()
obj.methodA()
obj.methodB()

"""Multi-level Inheritance: In multi-level inheritance, a subclass inherits from a parent class,
 which in turn inherits from another parent class."""

class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def print_details(self):
        print("Make:", self.make)
        print("Model:", self.model)

class Car(Vehicle):
    pass
```
